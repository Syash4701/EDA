# 13.`Is there a correlation between customer age and the likelihood of returning an order?


```python
df.columns
```

```python
df['OrderReturn'].value_counts()
```

```python
df['ReturnEncoded'] = [1 if r == True else 0 for r in df['OrderReturn']]
df[['ReturnEncoded', 'OrderReturn']].head()
```

```python
df['ReturnEncoded'].value_counts()
```

```python
age_return = {}
for age_group in df['AgeGroup'].unique():
  cust = df[df['AgeGroup']==age_group]
  #print(cust)
  count_return = sum(cust['ReturnEncoded'])
  #print(count_return)
  a= len(cust)
  print(a)
  probability = count_return/len(cust)
  age_return[age_group] = probability

#age_return
```

```python
sns.barplot(x=list(age_return.keys()), y= list(age_return.values()))
plt.title('Probability of Returning Order by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Probability')
```
